### Laboratorio de Escalonamento ###

Neste laboratório, iremos simular o funcionamento de algoritmos de escalonamento básicos para entender melhor seu funcionamento.

Na célula abaixo, temos uma classe Processo, que tem as informações de execução, e uma classe de exemplo de escalonamento apenas com os protótipos:

(orientação a objeto em Python)

```python
import numpy as np
import random

class Processo(object):
    def __init__(self,pnome,pio,ptam,prioridade,tempoChegada,bilhetes=0):
      self.nome = pnome
      self.io = pio # Probabilidade de fazer E/S, inicialmente zero
      self.tam = ptam # Quantos Timeslices sao necessarios para terminar
      self.prio = prioridade # Prioridade, eh desnecessaria agora
      self.tempo_prio = 0
      self.chegada = tempoChegada
      self.passos=0

    def roda(self,quantum=None): # se rodar sem quantum, o processor roda ate o fim
      if random.randint(1,100)<self.io: #Verifica se fez E/S
        self.tam-=1
        print(self.nome," fez e/s, falta ",self.tam)
        return 1, True #True que fez E/S


      if quantum is None or self.tam<quantum:
          quantum = self.tam
      self.tam -= quantum
      self.tempo_prio += quantum
      print(self.nome," rodou por ",quantum," timeslice, faltam ",self.tam)
      return quantum, False # False se nao fez E/S


class lottery(object):
    def __init__(self,vprontos=[]):
        self.prontos = vprontos 
        self.bilhetes = 0
    def pronto(self,Processo):
        if Processo is not None:
            value = Processo.prio
            self.bilhetes += value 
            self.prontos.append(Processo)
            self.prontos.sort(reverse=True,key=lambda x: x.prio)
        
    def proximo(self,ticket):
        if(len(self.prontos)>0):
            prob = 0
            for i in range(len(self.prontos)):
                prob+=self.prontos[i].prio
                if(ticket<=prob):
                    p = self.prontos[i]
                    return p
                
        else:
            return None
        
    def ended(self, Processo):
        if(Processo is not None):
            for i in range(len(self.prontos)):
                if(Processo.nome==self.prontos[i].nome):
                    self.bilhetes -= self.prontos[i].prio
                    del self.prontos[i]
                    break


class stride(object): 
    def __init__(self,vprontos=[]):
        self.prontos = vprontos 
        self.bilhetes = 0
        
    def pronto(self,Processo):
        if Processo is not None:
            value = Processo.prio
            self.bilhetes += value 
            self.prontos.append(Processo)
            self.prontos.sort(key=lambda x: x.passos)
        
    def proximo(self):
        if(len(self.prontos)>0):
            self.prontos[0].passos += int(100/self.prontos[0].prio)
            p = self.prontos[0]
            self.prontos = self.prontos[1:]
            return p
        else:
            return None
        
    def ended(self,Processo):
        if(Processo is not None):
            for i in range(len(self.prontos)):
                if(Processo.nome==self.prontos[i].nome):
                    self.bilhetes -= self.prontos[i].prio
                    del self.prontos[i]
                    break


class MLFQ(object): # Protótipo de escalonador de exemplo
    def __init__(self, fila_quantum, tempboost = None,vprontos=[]):
        self.tempo = 0
        self.quantum = fila_quantum
        self.numf = len(fila_quantum)
        self.boost = tempboost
        
        self.fila = [None]*self.numf
        
        for i in range(self.numf):
            self.fila[i] = []
        
        self.fila[0] = vprontos

    def pronto(self,pro):
        if pro.tempo_prio >= self.quantum[pro.prio]:
          pro.prio = min(self.numf-1, pro.prio+1)
          p.tempo_prio = 0

        self.fila[pro.prio].append(pro)

    def proximo(self):
        pro = None
        for i in range(len(self.fila)):
            if len(self.fila[i]) > 0:
                pro = self.fila[i][0]
                self.fila[i] = self.fila[i][1:]
                break
        return pro
    
    
    def sub(self, rodou):
      if self.boost != None:
          self.tempo += rodou
          if self.tempo >= self.boost:
              self.tempo = 0
              for i in range(1, self.numf):
                  for p in self.fila[i]:
                      p.tempo_prio = 0
                      p.prio = 0
                      self.fila[0].append(p)
                  self.fila[i] = []

    def prontos(self):
      for i in self.fila:
          if len(i) > 0:
              return 1
      return 0;    

        
```

* 2 - Existem duas configurações possíveis que transformam o MLFQ em um Round Robin. A primeira seria se tivessemos processos da mesma natureza e todos eles tivessem o mesmo tamanho, assim, mesmo que os processos trocassem de uma fila para outra, todos eles seriam executados em sequência igual ao RB. Técnicamente seriam vários round robins então podemos dizer que essa fórmula é "igual ao Round Robin". Teoricamente, a verdadeira fórmula que transforma o MLFQ em Round Robin seria um MLFQ com apenas uma fila, uma prioridade. Assim o MLFQ seria literalmente o Round Robin.
