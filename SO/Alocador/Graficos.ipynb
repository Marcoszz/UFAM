```python
import matplotlib.pyplot as plt
import glob
import sys
import pandas as pd
import numpy as np
```

```python
glob.glob("/home/marco/Desktop/UFAM/SO/Alocador/graficos.txt")

pd_best = pd.read_csv("graficos.txt",sep=' ', names = ["Número Espaços Livres","Soma Total dos Espaços","Média dos Espaços"])
pd_first = pd.read_csv("graficos1.txt",sep=' ', names = ["Número Espaços Livres","Soma Total dos Espaços","Média dos Espaços"])
pd_next = pd.read_csv("graficos2.txt",sep=' ', names = ["Número Espaços Livres","Soma Total dos Espaços","Média dos Espaços"])
```

```python
numero_best = []
total_best = []
media_best = []
numero_first = []
total_first = []
media_first = []
numero_next = []
total_next = []
media_next = []

for i in range(len(pd_best["Número Espaços Livres"])):
    numero_best.append(pd_best["Número Espaços Livres"][i])
    numero_first.append(pd_first["Número Espaços Livres"][i])
    numero_next.append(pd_next["Número Espaços Livres"][i])

for i in range(len(pd_best["Soma Total dos Espaços"])):
    total_best.append(pd_best["Soma Total dos Espaços"][i])
    total_first.append(pd_first["Soma Total dos Espaços"][i])
    total_next.append(pd_next["Soma Total dos Espaços"][i])
    
for i in range(len(pd_best["Média dos Espaços"])):
    media_best.append(pd_best["Média dos Espaços"][i])
    media_first.append(pd_first["Média dos Espaços"][i])
    media_next.append(pd_next["Média dos Espaços"][i])
    
numero_best = np.array(numero_best)
total_best = np.array(total_best)
media_bets = np.array(media_best)
numero_first = np.array(numero_first)
total_first = np.array(total_first)
media_first = np.array(media_first)
numero_next = np.array(numero_next)
total_next = np.array(total_next)
media_next = np.array(media_next)
```

```python
plt.figure(figsize=(15,10))

plt.title('Número de Espaços Livres x Iterações')
l1, = plt.plot(range(1,len(numero_best)+1), numero_best,"yellowgreen") 
l2, = plt.plot(range(1,len(numero_best)+1), numero_first,"darkturquoise")
l3, = plt.plot(range(1,len(numero_best)+1), numero_next,"tomato")
plt.legend([l1,l2,l3],['Best-fit','First-fit','Next-fit'])
plt.xlabel("Iterações")
plt.ylabel("Quantidade de Espaços")
plt.savefig("numero.png")


plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Best-fit: Soma Total de Espaços Livres x Iterações')
plt.plot(range(len(numero_best)), total_best,"yellowgreen",label="Best-fit") 
plt.yscale("log")
plt.xlabel("Iterações")
plt.ylabel("Soma Total de Espaços")
plt.legend()
plt.savefig("soma_best.png")
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.title('First-fit: Soma Total de Espaços Livres x Iterações')
plt.plot(range(len(numero_best)), total_first,"darkturquoise",label="First-fit")
plt.yscale("log")
plt.xlabel("Iterações")
plt.ylabel("Soma Total de Espaços")
plt.legend()
plt.savefig("soma_first.png")
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.title('Next-fit: Soma Total de Espaços Livres x Iterações')
plt.plot(range(len(numero_best)), total_next,"tomato",label="Next-fit")
plt.yscale("log")
plt.xlabel("Iterações")
plt.ylabel("Soma Total de Espaços")
plt.legend()
plt.savefig("soma_next.png")
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Best-fit: Média de Espaços Livres x Iterações')
plt.plot(range(len(media_best)), media_best,"yellowgreen",label="Best-fit") 
plt.yscale("log")
plt.xlabel("Iterações")
plt.ylabel("Média de Espaços")
plt.legend()
plt.savefig("media_best.png")
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('First-fit: Média de Espaços Livres x Iterações')
plt.plot(range(len(media_best)), media_first,"darkturquoise",label="First-fit")
plt.yscale("log")
plt.xlabel("Iterações")
plt.ylabel("Média de Espaços")
plt.legend()
plt.savefig("media_first.png")

plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Next-fit: Média de Espaços Livres x Iterações')
plt.plot(range(len(media_best)), media_next,"tomato",label="Next-fit")
plt.yscale("log")
plt.xlabel("Iterações")
plt.ylabel("Média de Espaços")
plt.legend()
plt.savefig("media_next.png")
plt.show()
```
