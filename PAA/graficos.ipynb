```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# C++
nchaves = [10,100,1000,10000,100000,1000000,10000000,100000000]

memoHash = [15600,15600,15600,52800,4400000,41700000,397700000,3700000000]
tempHash = [3.3e-05,0.000223,0.001259,0.011631,0.032698,0.280289,2.6829,26.181]


memoAVL = [15600,15600,15600,52800,5900000,57000000,550400000,7400000000]
tempAVL = [2.2e-05,0.0002,0.002269,0.013268,0.067884,0.792385,9.35265,59.862]


plt.figure(figsize=(15,5))
plt.rc('font',size=18)
plt.plot(nchaves,tempHash,'*--',label="Tempo Hash")
plt.plot(nchaves,tempAVL,'*--',label="Tempo RB")
plt.title("Tempo x Nmero de Chaves C++")
plt.xscale("log")
plt.xlabel("Chaves")
plt.ylabel("Tempo(s)")
plt.rc('font',size=16)

plt.legend()
plt.savefig('tempoc++.png', format='png')
plt.show()

plt.figure(figsize=(15,5))
plt.rc('font',size=18)
plt.plot(nchaves,memoHash,'*--',label="Memo Hash")
plt.plot(nchaves,memoAVL,'*--',label="Memo RB")
plt.title("Memria x Nmero de Chaves C++")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Chaves")
plt.ylabel("Memria")
plt.rc('font',size=16)

plt.legend()
plt.savefig('memoc++.png', format='png')
plt.show()
```
