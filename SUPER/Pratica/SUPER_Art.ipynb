```python
# Reading all the data

names = ['drive01','drive02','drive03','drive04','drive05','drive06',
         'drive07','drive08','drive09','drive10','drive11','drive12',
        'drive13','drive14','drive15','drive16','drive17a','drive17b']

records = []

for i in range(0,len(names)):
    record = wfdb.rdrecord(path[0]+names[i])
    records.append(record)

wfdb.plot_wfdb(records[0])
display(records[0].__dict__)
```

```python
for i in range(0,len(records)):    
    columns = records[i].sig_name
    df = pd.DataFrame(records[i].p_signal,columns=columns)
    df.index = [(1 / 15.5) * i for i in range(len(df))]
    df.index = pd.to_datetime(df.index,unit='s')
    df.to_csv(names[i]+'.txt', sep = ' ')
    
    
```
